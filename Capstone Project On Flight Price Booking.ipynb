# Convert all the attributes into integer data type which are of object data type except Flight column.
 


```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df["airline"]=label.fit_transform(df["airline"])
df["source_city"]=label.fit_transform(df["source_city"])
df["departure_time"]=label.fit_transform(df["departure_time"])
df["stops"]=label.fit_transform(df["stops"])
df["arrival_time"]=label.fit_transform(df["arrival_time"])
df["class"]=label.fit_transform(df["class"])
df["destination_city"]=label.fit_transform(df["destination_city"])
```

```python
df.info()
```

# drop flight column


```python
df.drop(columns=["flight"],inplace=True,axis=0)
df
```

# checking for multicolinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list=[]
for col in df.columns:
    if((df[col].dtype!="object")&(col!="price")):
        col_list.append(col)
col_list
```

```python
x=df[col_list]
vif_data=pd.DataFrame()
vif_data["feature"]=x.columns
vif_data["vif"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
print(vif_data)
```

from values of vif we can say input features are not corelated to each other

# Indepenedent And Dependent features

```python
X=df.iloc[:,:-1]
X
y=df["price"]
y
```

# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
X_train
```

# Standardadizing the dataset

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

# Model Accuracy Of  linear regression

```python
# R2 score
```

```python
from sklearn.metrics import *
r2_score(y_test,y_pred)
```

```python
# Mean squared error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
# Mean Absolute Error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
sns.distplot(y_pred,label="Predicted")
sns.distplot(y_test,label="Actual")
plt.title("Linear Regression")
```

Actual and predicted value dont match at all

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,y_train)
dt_pred=dt.predict(X_test)
```

# Model Accuracy Of Decision Tree Regressor

```python
from sklearn.metrics import *
r2_score(y_test,dt_pred)
```

```python
mean_squared_error(y_test,dt_pred)
```

```python
mean_absolute_error(y_test,dt_pred)
```

```python
sns.distplot(y_test,label="Actual")
sns.distplot(dt_pred,label="predicted")
plt.title("Decision Tree Regressor")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
rf_pred=rf.predict(X_test)
```

# Model Accuracy Of Random Forest Regressor

```python
from sklearn.model_selection import *
r2_score(y_test,rf_pred)
```

```python
mean_squared_error(y_test,rf_pred)
```

```python
from math import sqrt
rmse=sqrt(mean_squared_error(y_test,rf_pred)) 
rmse
```

```python
mean_absolute_error(y_test,rf_pred)
```

```python
sns.distplot(y_test,label="Actual")
sns.distplot(rf_pred,label="Predicted")
plt.title("Random Forest Regressor")
```

# CONCLUSION=Model Accuracy for a given dataset is most accurate for Random Forest Regressor
